# Q6. Clean the data: In the dataset, there are a lots of titles in Name feature. Remap the tile to: 0 for "Mr", 1 for "Miss", and 3 for the left.

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2,
                "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
df['Title'] = df['Title'].map(title_mapping)
df.head()
```

# Q7. Remap the sex to use int value for easier usage later: 0 for Male, and 1 for female

```python
sex_mapping = {"male": 0, "female": 1}
df['Sex'] = df['Sex'].map(sex_mapping)
df.head()
```

# Q8. Try grouping the age: 0-16 y.o to group 0, 16-26 to group 1, 26-36 to group 2, 36-62 to group 3, and the left to group 4

```python
df['Age_range'] = pd.cut(df['Age'], bins=[0, 16, 26, 36, 62, 300], labels=[0, 1, 2, 3, 4])
df.drop('Age', axis=1)
```
